```python
duplicates_exist = cleaned_bnb.duplicated().sum()
print("Number of duplicated rows:", duplicates_exist)

if duplicates_exist > 0:
    print("Proof: At least one record has been duplicated.")
else:
    print(" No duplicate records found.")
```

```python
bnb_cleaned = cleaned_bnb.drop_duplicates()
bnb_cleaned
```

```python
listings_count = bnb_cleaned.groupby('neighbourhood group').size().reset_index(name='Number_of_Listings')
listings_count
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=listings_count, x='neighbourhood group', y='Number_of_Listings')
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.show()
```

```python
pol_data = pd.read_csv("C:\\Users\\DIGITALDREAM CYBER\\Downloads\\Politics_Tweets.csv")
pol_data
```

```python
import re

def extract_urls(media_str):
    # If cell is NaN or not a string, return None
    if pd.isna(media_str):
        return None
    # Find all URLs in the string
    urls = re.findall(r'https://pbs\.twimg\.com/media/[A-Za-z0-9_-]+\.\w+', media_str)
    # Join multiple URLs with commas
    return ', '.join(urls) if urls else None
```

```python
pol_data['media'] = pol_data['media'].apply(extract_urls)
pol_data.head()
```
